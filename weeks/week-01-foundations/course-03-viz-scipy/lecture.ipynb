## 3. SciPy — probability

### Plot three Gaussians on one axes

```python
x = np.linspace(-4, 4, 200)
for mu, sigma in [(0, 1), (0, 0.5), (1, 1)]:
    plt.plot(x, stats.norm.pdf(x, mu, sigma), label=f'N({mu}, {sigma})')
plt.legend(); plt.title('Normal PDFs'); plt.show()
```

### Fit a Gaussian to data

```python
rng1 = np.random.default_rng(1)
samples = rng1.normal(loc=2.0, scale=1.5, size=1000)
mu_hat, sigma_hat = stats.norm.fit(samples)
print(f'mu_hat = {mu_hat:.3f}, sigma_hat = {sigma_hat:.3f}')

xs = np.linspace(samples.min(), samples.max(), 200)
plt.hist(samples, bins=30, density=True, alpha=0.5)
plt.plot(xs, stats.norm.pdf(xs, mu_hat, sigma_hat), lw=2, label='fitted')
plt.legend(); plt.show()
```

## 4. SciPy — optimization

### The 1-D loss landscape

```python
f = lambda x: (x - 3)**2 + np.sin(5 * x)
xs = np.linspace(-2, 8, 400)
plt.plot(xs, f(xs))
plt.title('A loss surface'); plt.xlabel('x'); plt.ylabel('loss'); plt.show()
```

### Minimize — this is what models do

```python
res = optimize.minimize(f, x0=0.0)
print('argmin =', res.x[0])
print('min    =', res.fun)
print('iters  =', res.nit)
```

```python
plt.plot(xs, f(xs))
plt.axvline(res.x[0], color='red', ls='--', label=f'min at x={res.x[0]:.3f}')
plt.legend(); plt.show()
```

### Local minima — initialization matters

```python
res2 = optimize.minimize(f, x0=10.0)
print('from x0=0  :', res.x[0])
print('from x0=10 :', res2.x[0])
```

### Takeaway
Every supervised model: define a loss `L(parameters)` and call an optimizer. Today you saw the whole loop in 3 lines.

### Exercise 3 — Solution

# Exercise 3 — Grand Finale Solution

```python
import sys, pathlib
p = pathlib.Path.cwd()
while not (p / 'pyproject.toml').exists() and p != p.parent:
    p = p.parent
if str(p) not in sys.path:
    sys.path.insert(0, str(p))
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from shared.data_utils import load_dataset
rng = np.random.default_rng(0)

df = load_dataset('stock-prices')
PRICE = df.columns[1]   # second column = closing price
DATE = df.columns[0]
mask = rng.random(len(df)) < 0.05
df.loc[mask, PRICE] = np.nan
print('shape:', df.shape, ' price col:', PRICE)
```

**Step 1.** Inspect.

```python
print('NaNs in price:', df[PRICE].isna().sum())
print('date dtype  :', df[DATE].dtype)
```

**Step 2.** Parse + sort.

```python
df[DATE] = pd.to_datetime(df[DATE])
df = df.sort_values(DATE).reset_index(drop=True)
df.head()
```

**Step 3.** Forward-fill.

```python
df[PRICE] = df[PRICE].ffill()
assert df[PRICE].isna().sum() == 0
```

**Step 4.** 7-day moving average with NumPy.

```python
window = 7
kernel = np.ones(window) / window
ma = np.convolve(df[PRICE].values, kernel, mode='valid')
# Pad the front so lengths line up for plotting
ma_full = np.concatenate([np.full(window - 1, np.nan), ma])
print('ma len:', len(ma_full), '  df len:', len(df))
```

**Step 5.** Plot.

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(df[DATE], df[PRICE], label='price', alpha=0.5)
ax.plot(df[DATE], ma_full, label='7-day MA', lw=2)
ax.set_xlabel('date'); ax.set_ylabel(PRICE); ax.set_title('AAPL closing price + 7-day MA')
ax.legend(); fig.autofmt_xdate();
```

**Task 3.**

```python
samples = rng.normal(size=10_000)
print('mean:', samples.mean(), '  std:', samples.std())
assert abs(samples.mean()) < 0.05
assert abs(samples.std() - 1) < 0.05
```

```python
n = 3
ix = np.arange(n)
I = (ix[:, None] == ix[None, :]).astype(int)
print(I)
assert np.array_equal(I, np.eye(n, dtype=int))
```
